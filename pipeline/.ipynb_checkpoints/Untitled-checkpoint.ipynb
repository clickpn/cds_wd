```python
import pandas as pd
import re
```

```python
df = pd.read_csv('./csv_file/abalone.csv')
```

```python
high_kws_lst = [
'black tail', 'mule', 'roe deer',
'grey fox',
'ivory',
'geese']

low_kws_lst = [
'polished handicraft abalone',
'blacktail deer',
'quail', 'duck farm raised',
'bobcat',
'hybrid wolf',
'deer hide',
'white tail deer',
'red fox'
'sting ray'
]
```

```python
def adjust_priority(data):
    data['adj_scores'] = data['rel_scores']
    for index, row in data.iterrows():
        score = row['rel_scores']
        if score > 0.75 and score <= 1:
            if any(word in row for word in high_kws_lst):
                row['adj_scores'] = 1
            else:
                row['adj_scores'] = score
        elif score <= 0.75 and score > 0.5:
            if any(word in row for word in low_kws_lst):
                row['adj_scores'] = 0.6
            else:
                row['adj_scores'] = score
        else:
            row['adj_scores'] = score
    return data
```

```python
adjust_priority(df)
```
